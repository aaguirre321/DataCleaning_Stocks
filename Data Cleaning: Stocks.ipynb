```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
```

<b> Import Stocks Data <b>

```python
url = "https://raw.githubusercontent.com/PhilChodrow/PIC16A/master/discussion/timeseries-sample-data.csv"

df = pd.read_csv(url)
dates = np.array(pd.to_datetime(df["dates"]))
symbols = np.array(df["symbols"])
prices = np.array(df["prices"])
```

<b>Describe Stocks Data<b>

```python
def describe(n, dates, symbols, prices):
    ix = np.array(symbols == n)
    true_count = np.count_nonzero(ix)
    print("There are {} entries for symbol {}".format(true_count, n))
    print("The average value was {}".format(np.mean(prices[ix])))
    print("The max value was {}".format(np.max(prices[ix])))
    print("The time period ran from {} to {}.".format(np.min(dates[ix]), np.max(dates[ix])))
```

```python
describe(0, dates, symbols, prices)
```

<b> Data Plot Prior to Cleaning<b>

```python
fig, axarr = plt.subplots(3, 1, figsize = (8, 8))
def plot_symbol(s, dates, symbols, prices, ax):
    ix = np.array(symbols == s)
    axarr[ax].plot(dates[ix], prices[ix])
    return fig
for i in range(3):
    plot_symbol(i,dates,symbols,prices,i)
```

<b> Sorted Data <b>

```python
def plot_symbol(s, dates, symbols, prices, ax):
    ix = np.array(symbols == s)
    iy = np.argsort(dates[ix])
    axarr[ax].plot(dates[ix][iy], prices[ix][iy])
```

```python
fig, axarr = plt.subplots(3, 1, figsize = (8, 8))
plot_symbol(0, dates, symbols, prices, 0)
plot_symbol(1, dates, symbols, prices, 1)
plot_symbol(2, dates, symbols, prices, 2)
```
